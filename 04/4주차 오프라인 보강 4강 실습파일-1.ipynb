#### 원하는 컬럼 선택하기

```python
basket_ball.head()
```

#### 조건을 주어 원하는 컬럼 선택하기
- & : and 둘 다 만족
- | : or 둘 중 하나만 만족해도 됨

```python
basket_ball[basket_ball['3P'] > 0.5]
```

```python
basket_ball[(basket_ball['3P'] > 0.5) & (basket_ball['2P'] > 2.0)]
```

#### 원하는 행 선택하기

- .loc, .iloc로 row 선택 가능
- loc - 인덱스 자체를 사용
- iloc - 0 based index로 사용


```python
basket_ball.head()
```

```python
basket_ball.loc[0]
```

```python
basket_ball.loc[[0]]
```

```python
basket_ball.iloc[0]
```

```python
basket_ball.iloc[[0]]
```

```python
basket_ball.iloc[[0, 3, 6]]
```

```python
basket_ball.loc[[0, 3, 6]]
```

#### 행과 열 동시 선택하기

```python
basket_ball.loc[[0,3,6],['Player', 'TRB', 'BLK']]
```

```python
basket_ball.iloc[[0,3,6],[0,4,7]]
```
