#### 새 column 추가하기

```python
#리스트에서 새로운 데이터 추가히기
list = [0,1,2,3,4,5]
list[0] = 'a'
list
```

```python
basket_ball.head()
```

```python
basket_ball['3P_double'] = 0
basket_ball.head()
```

```python
basket_ball['3P_double'] = basket_ball['3P'] * 2
basket_ball
```

```python
basket_ball['3P_tripple'] = basket_ball['3P_double'] + basket_ball['3P']
basket_ball
```

```python
basket_ball.insert(3, '3P10', basket_ball['3P']*10)
basket_ball.head()
```

#### 새 row 추가하기

```python
basket_ball.loc['new_row'] = 0
basket_ball.tail()
```

```python
basket_ball.loc['new_row'] = basket_ball.loc['new_row'] + basket_ball.loc[199]
basket_ball.tail()
```

#### column 삭제하기
- drop 함수 사용하여 삭제, axis = 1
- 원본을 삭제하기 위해서는 inplace = True 매개변수를 주어야 함
- 리스트를 사용하여 멀티플 삭제 가능

```python
basket_ball.drop('3P_tripple', axis=1)
```

```python
basket_ball.head()
```

```python
basket_ball.drop('3P_tripple', axis=1, inplace=True)
basket_ball.head()
```

```python
basket_ball.drop(['3P10','3P_double'], axis=1, inplace=True)
basket_ball.head()
```

#### row 삭제하기
- drop 함수 사용하여 삭제, axis = 0, 
- 원본을 삭제하기 위해서는 inplace = True 매개변수를 주어야 함
- 리스트를 사용하여 멀티플 삭제 가능

```python
basket_ball.drop([0], axis = 0)
```

```python
basket_ball.head()
```

```python
basket_ball.drop([0], axis = 0, inplace=True)
basket_ball.head()
```

```python
basket_ball.drop([1,2,3], axis = 0, inplace=True)
basket_ball.head()
```

```python
basket_ball.head()
```

#### 컬럼 간 상관관계 계산하기
- corr 함수를 통해 상관계수 연산(-1, 1 사이의 결과)
 - -1에 가까울 수록 음의 상관관계
 - 1에 가까울 수록 양의 상관관계
 - 0이면 관계없음

```python
basket_ball.corr()
```

```python
import matplotlib.pyplot as plt

plt.matshow(basket_ball.corr())  #색이 밝을 수록 더 관계가 깊다
```

```python
plt.scatter(basket_ball['TRB'],basket_ball['2P'])
plt.xlabel('TRB')
plt.ylabel('2P')
```

```python
plt.scatter(basket_ball['3P'],basket_ball['2P'])
plt.xlabel('3P')
plt.ylabel('2P')
```
